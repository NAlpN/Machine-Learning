```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

data = pd.read_csv("polynomial.csv",sep = ";")
```

```python
data
```

```python
plt.scatter(data['deneyim'], data['maas'])
plt.xlabel('Deneyim (yıl)')
plt.ylabel('Maaş (aylık)')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(data[['deneyim']],data['maas'])

plt.xlabel('Deneyim (yıl)')
plt.ylabel('Maaş (aylık)')

plt.scatter(data['deneyim'],data['maas'])

xekseni = data['deneyim']
yekseni = reg.predict(data[['deneyim']])
plt.plot(xekseni, yekseni, color='green', label = "Tahminler")
plt.legend()
plt.show()
```

```python
pol_reg = PolynomialFeatures(degree = 4)
x_pol = pol_reg.fit_transform(data[['deneyim']])
```

```python
reg = LinearRegression()
reg.fit(x_pol,data['maas'])
```

```python
y_head = reg.predict(x_pol)
plt.plot(data['deneyim'],y_head, color='red', label='Tahminler')

plt.legend()

plt.scatter(data['deneyim'],data['maas'])

plt.show()
```

```python
x_pol_deneme = pol_reg.fit_transform([[4.5]])
reg.predict(x_pol_deneme)
```

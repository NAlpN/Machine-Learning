```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

```python
veriler = pd.read_csv('maaslar.csv')

x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:]
X = x.values
Y = y.values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(x,lin_reg.predict(X), color = 'blue')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(X)
```

```python
x_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
```

```python
x_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.show()
```

```python
lin_reg.predict([[11]])
```

```python
lin_reg.predict([[6.6]])
```

```python
lin_reg2.predict(poly_reg.fit_transform([[6.6]]))
```

```python
lin_reg2.predict(poly_reg.fit_transform([[11]]))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc1=StandardScaler()
```

```python
x_olcekli = sc1.fit_transform(X)
```

```python
sc2=StandardScaler()
```

```python
y_olcekli = np.ravel(sc2.fit_transform(Y.reshape(-1,1)))
```

```python
from sklearn.svm import SVR
```

```python
svr_reg = SVR(kernel='rbf')
svr_reg.fit(x_olcekli,y_olcekli)
```

```python
plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_reg.predict(x_olcekli),color='blue')

plt.show()
```

```python
svr_reg.predict([[11]])
```

```python
svr_reg.predict([[6.6]])
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
r_dt = DecisionTreeRegressor(random_state=0)
r_dt.fit(X,Y)
Z = X + 0.5
K = X - 0.4
```

```python
plt.scatter(X,Y, color='red')
plt.plot(x,r_dt.predict(X), color='blue')
plt.plot(x,r_dt.predict(Z),color='green')
plt.plot(x,r_dt.predict(K),color='yellow')
plt.show()
```

```python
r_dt.predict([[11]])
```

```python
r_dt.predict([[6.6]])
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_reg=RandomForestRegressor(n_estimators = 10,random_state=0)
rf_reg.fit(X,Y.ravel())
```

```python
rf_reg.predict([[6.6]])
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,rf_reg.predict(X),color='blue')

plt.plot(X,rf_reg.predict(Z),color='green')
plt.plot(x,r_dt.predict(K),color='yellow')
plt.show()
```

```python
print('Linear R2 degeri')
r2_score(Y, lin_reg.predict(X))
```

```python
print('Polynomial R2 degeri')
r2_score(Y, lin_reg2.predict(poly_reg.fit_transform(X)))
```

```python
print('SVR R2 degeri')
r2_score(y_olcekli, svr_reg.predict(x_olcekli))
```

```python
print('Decision Tree R2 degeri')
r2_score(Y, r_dt.predict(X))
```

```python
print('Random Forest R2 degeri')
r2_score(Y, rf_reg.predict(X))
```
